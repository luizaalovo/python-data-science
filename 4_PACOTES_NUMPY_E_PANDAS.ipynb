O DataFrame apresenta 8 colunas, todas preenchidas, sem valores nulos. Os dados se referem a alunos de uma escola. Abaixo é possível verificar os dados que cada uma das colunas apresenta:

* Gender: esta coluna apresenta o gênero da pessoa, separando-o em dois tipos:
 -  Male:  Masculino
 -  Female: Feminino

* Race/ethnicity:  esta columa apresenta a etnia da pessoa, separando-a em grupos
  - 'Grupo A'
  -  'Grupo B'
  - 'Grupo C'
  - 'Grupo D'
  - 'Grupo E'

* Parental level of education -> Esta coluna apresenta o grau de escolaridade dos pais dos alunos, separando-os em seis grupos:
  - High school: Ensino médio
  - Some high school: Estudante do ensido médio
  - Some college: Universitário
  - Associate's degree: Curso de duração média completo
  - Bachelor's degree: Bacharelado
  - Master's degree: Mestrado

* Lunch -> Esta coluna apresenta o tipo de refeição disponibilizada ao aluno, separando-a em dois tipos:
  - Standard: refeição padrão
  - Free/reduce: refeição pode ser livre ou com restrições

* Test preparation course -> Esta coluna apresenta informação se o aluno já fez o curso preparatório, separando-o em dois tipos:
  - Completed: caso o aluno já tenha realizado o curso
  - None: caso o aluno ainda não tenha realizado o curso

* Math score: Pontuação em matemática

* Reading score: Pontuação em leitura

* Writing score: Pontuação em escrita


# 2. Utilize a função _where_ do pacote _NumPy_ e substitua os valores da coluna **"test preparation course"** para 0, quando _none_ e 1 quando _completed_. Use a mesma função para substituir os valores da coluna **"gender"** para M ou F.

```python
#Substituindo dados da coluna 'test preparation course'
dados["test preparation course"] = np.where(dados["test preparation course"] == "completed", 1, 0)

# Substituindo dados da coluna 'gender'
dados["gender"] = np.where(dados["gender"] == "male", "M", "F")

dados.head(20)
```

#### 3. Faça a soma das notas para cada aluno (ou seja, soma linha a linha). Aramazene o resultado em uma nova coluna com o nome **"total score"**.

```python
dados['total score'] = (dados['math score'] + dados['reading score'] + dados['writing score'])
dados.head()
```

#### 4. Plote um gráfico do tipo histograma da nova variável **"total score"**. Utilize o método _plot_ do próprio Pandas.

#### 5. Crie 3 grupos distintos: um com o público que teve **total score** menor ou igual a 150, outro que teve **total score** maior que 150, mas menor ou igual a 250 e, por fim, quem teve **total score** acima de 250. Mostre a dimensão de cada um dos 3 _datasets_ criados.

```python
grupo1 = dados.loc[dados["total score"] <= 150]
grupo2 = dados.loc[(dados["total score"] > 150) & (dados["total score"] <= 250)]
grupo3 = dados.loc[dados["total score"] > 250]

# Exibindo as dimensões de cada grupo
print("Dimensão do Grupo 1 (total score <= 150):", grupo1.shape[0])
print("Dimensão do Grupo 2 (150 < total score <= 250):", grupo2.shape[0])
print("Dimensão do Grupo 3 (total score > 250):", grupo3.shape[0])
```

#### 6. Para os 3 novos _datasets_ criados, mostre a contagem de homems e mulheres e o percentual dos valores da variável **"race/ethnicity"**.

```python
dados["group"] = np.where(dados["total score"] <= 150, "Grupo 1 (<=150)",
                       np.where(dados["total score"] <= 250, "Grupo 2 (150-250)",
                                "Grupo 3 (>250)"))

cont_gen = dados.groupby("group")["gender"].value_counts()
perc_race = dados.groupby("group")["race/ethnicity"].value_counts(normalize=True) * 100

print("Contagem de homens e mulheres por grupo:\n", cont_gen)
print("\nPercentual de raça/etnia por grupo:\n", perc_race)
```
